```python
#teğet ile bulma yöntemi
a=1
b=-7
c=10
i=1
def f(x):
    return (a*x**2 + b * x + c)
def ft(x):
    return (2*a*x + b)

xr=int(input("tahmini girin: "))
while(f(xr)!=0):
    i=i+1
    xr=xr-(f(xr)/ft(xr))
    print(i,".hamle",xr)
print(xr)
```
